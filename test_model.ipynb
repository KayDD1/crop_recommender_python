```python
import pickle
with open('nb_model.pkl', 'rb') as f:
  nb_new = pickle.load(f)
```

```python
with open('scaled.pkl', 'rb') as f:
  scaled_new = pickle.load(f)
```

```python
import pandas as pd

new_df = pd.read_csv("sythetic_crop_data.csv")
new_df.head()
```

```python
new_df.drop('Unnamed: 0', axis=1, inplace=True)
new_df.dropna(inplace=True) 
new_df.shape
```

```python
from scipy.stats import boxcox

opt_lambda = boxcox(new_df['K'])[1]
transformed_column = boxcox(new_df['K'], lmbda=opt_lambda)
transformed_column
```

```python
new_df['K_new'] = transformed_column
new_df.drop('K', axis=1, inplace=True)
new_df.head(50)
```

```python
X = new_df.drop(['label'], axis=1)
y = new_df['label']
```

```python
x_scaler = scaled_new.fit_transform(X)
```

```python
x_scaler
df_scaled = pd.DataFrame(x_scaler, columns=X.columns)
df_scaled
```

```python
df_scaled.isnull().sum()
df_scaled.dropna(inplace=True)
```

```python
y_pred_new = nb_new.predict(df_scaled)
```

```python
y_pred_new
```

```python
df_scaled['Predictions'] = y_pred_new
```

```python
df_scaled.head(50)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y, y_pred_new)
print('Model accuracy: {0:.2f}%'.format(accuracy * 100))
```
